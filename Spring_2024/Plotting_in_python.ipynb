# Generalities

Plotting in python is relatively simple given that we have so many libraries that can help us. In our case we will discuss how to use matplotlib to generate simple plots of functions.

## Importing libraries

We will learn how to use two libraries:

<ol>
    <li><b>Numpy</b>: this library has a lot of mathematical functions and data structures that are very useful. </li>
    <li><b>Matplotlib</b>: this library can generate plots of data. </li>
</ol>

```python
import numpy as np
import matplotlib.pyplot as plt
```

The first line here import the <b>numpy</b> library and also defines the nickname <b>np</b> to refer to it, this nickname is completely free and we can use whatever we want for it. It has become standard to use np for numpy and so you might see it quite often if you read some code.

The second line imports a sub-library (though these are technically called modules) of matplotlib, namely <b>pyplot</b> and it also defines the nickname <b>plt</b> for it. It has also become standard to use plt as nickname for matplotlib.pyplot.

## Plotting shapes

To plot we can use the <b>plot</b> function, which is defined inside of pytplot, so we can use it by simply calling <b>plt.plot</b>. This function expects to recieve a list of x-coordinates and a list of y-coordinates of the points that we wish to plot.

### Example #1: Triangles


```python
vertices_xs = np.array([0,1,-1,0])
vertices_ys = np.array([0,3,3,0])

plt.plot(vertices_xs,vertices_ys)
plt.show()
```

The plot function draws straight lines joining the points provided, this is the reason why we get a triangle here. Note that we need to repeat one of the verices in order to close the triangle shape, in our case we repeat the vertex at (0,0)

## Example #2: Parabolas

I mentioned that we can draw curved shapes by using some non-linear functions, for example to plot a parabola we can do so by filling the array of y-values with a quadratic function of the x-values.

```python
quadratic_function = lambda a,b,c,x : a*x**2 + b*x + c
```

This python lambda corresponds to the function:

$$ f(x) = ax^2+bx+c $$

Different values of a,b, and c correspond to different parabolas, for instance let's see one of these:

```python
# Here we define parabola1 which is x^2+1
parabola1 = lambda x : quadratic_function(1,0,1,x)

# Here we define how many points we are going to use to plot
num_points = 100

# This line defines the x-values, can you see why the x-axis goes from -1 to 1 
# in the plot?
parabola1_xs = np.array([ -1+(2)*(float(i)/num_points) for i in range(num_points+1)])

# In this line we apply the function parabola1 to the x's to build the y's
parabola1_ys = np.array([ parabola1(xs) for xs in parabola1_xs ])

plt.plot(parabola1_xs,parabola1_ys)
plt.show()
```
